```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import odeint
g = 9.8
ell = 0.5
th0 = 20.*np.pi/180
def deriv(theta_current,t):
    g = 9.8
    ell - 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([dtheta,domega])
```

```python
thetas = [0,np.pi/4.,np.pi/2.,3*np.pi/4,np.pi]
for a in thetas:
    t = np.arange(0.,10.+0.01,0.01)
    omega0 = a
    theta_init = np.array([th0,omega0])
    theta = odeint(deriv,theta_init,t)
    plt.plot(t,theta[:,0], label = ("Omega 0 = " + str(a)))
plt.grid()
plt.ylim(-1,3)
plt.legend(loc = "upper left")
```
