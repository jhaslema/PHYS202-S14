```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import odeint
def deriv(theta_current,t):
    g = 9.8
    ell = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([dtheta,domega])
```

```python
thetas = [0,np.pi/4.,np.pi/2.,3*np.pi/4,np.pi]
for a in thetas:
    t = np.arange(0.,50.+0.01,0.01)
    omega0 = 0
    th0 = a
    theta_init = np.array([th0,omega0])
    theta = odeint(deriv,theta_init,t)
    plt.plot(t,theta[:,0], label = ("Theta 0 = " + str(a)))
plt.grid()
plt.ylim(-20,5)
plt.xlim(0,25)
plt.xlabel("Time(seconds)",fontsize = 15)
plt.ylabel("Theta(Radians)",fontsize = 15)
plt.title("Theta vs Time",fontsize = 20)
plt.legend(loc = "lower left",fontsize = 10)
```

```python
tmax = 1
dx = .02
nsteps = int(tmax/dx)+1
x=np.zeros(nsteps)
y=np.zeros(nsteps)
x[0]=0
y[0]=0
for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    y[i+1] = y[i] + dx * (x[i] + 2 * y[i])
y[-1]
```

```python
#Euler Cromer Method .25 step size
tmax = 1
dx = .25
nsteps = int(tmax/dx)+1
x4=np.zeros(nsteps)
y4=np.zeros(nsteps)
v4=np.zeros(nsteps)
x4[0]=0
y4[0]=0
for i in range(len(x4)-1):
    x4[i+1] = x4[i] + dx
    v4[i+1] = x4[i] + 2*y4[i]
    y4[i+1] = y4[i] + dx * v4[i+1]
y4[-1]
```

```python
#Euler Cromer Method .02 step size
tmax = 1
dx = .02
nsteps = int(tmax/dx)+1
x5=np.zeros(nsteps)
y5=np.zeros(nsteps)
v5=np.zeros(nsteps)
x5[0]=0
y5[0]=0
for i in range(len(x5)-1):
    x5[i+1] = x5[i] + dx
    v5[i+1] = x5[i] + 2*y5[i]
    y5[i+1] = y5[i] + dx * v5[i+1]
y5[-1]
```

```python
#Midpoint Method .25 step size
tmax = 1
dx =.25
nsteps = int(tmax/dx)+1
x6=np.zeros(nsteps)
y6=np.zeros(nsteps)
v6=np.zeros(nsteps)
x6[0]=0
y6[0]=0
for i in range(len(x6)-1):
    x6[i+1] = x6[i] + dx
    v6[i+1] = x6[i] + 2*y6[i]
    y6[i+1] = y6[i] + 1/2.*(v6[i]+v6[i+1])*dx
y6[-1]
```

```python
#Midpoint Method .02 step size
tmax = 1
dx =.02
nsteps = int(tmax/dx)+1
x7=np.zeros(nsteps)
y7=np.zeros(nsteps)
v7=np.zeros(nsteps)
x7[0]=0
y7[0]=0
for i in range(len(x7)-1):
    x7[i+1] = x7[i] + dx
    v7[i+1] = x7[i] + 2*y7[i]
    y7[i+1] = y7[i] + 1/2.*(v7[i]+v7[i+1])*dx
print x7[-1]
```
